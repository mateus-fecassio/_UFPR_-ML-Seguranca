```python
import sys
import json
import numpy as np
import pandas as pd

pd.set_option('display.max_rows', 25)
pd.set_option('display.max_columns', 200)
```

```python
# Leitura dos arquivos dos protocolos escolhidos
df1 = pd.read_pickle('53_UDP_Dns.pkl', compression='bz2')
df2 = pd.read_pickle('80_TCP_HTTP.pkl', compression='bz2')
df = pd.concat([df1,df2])
```

```python
#Distribuição das classes escolhidas
df.groupby('ApplicationProtocolName')[['ApplicationProtocolNameFull']].count()
```

```python
df = df.drop(columns={'ApplicationProtocolNameFull'})

df.insert(0,'index',[ i for i in range(0,len(df))])
df.set_index('index', inplace=True)
```

```python
print(len(df.columns))
df.describe()
```

```python
for columns in df.columns:
    if ( str(df[columns].dtypes) == 'float64'):
        if ( (min(df[columns]) == 0.0) and (max(df[columns]) == 0.0) ):
            df = df.drop(columns=columns)
    
    elif ( str(df[columns].dtypes) == 'object'):
        temporary = df[columns].unique()
        if (temporary[0] == None):
            df = df.drop(columns=columns)
```

```python
print(len(df.columns))
```

```python
# Ajusta as colunas selecionadas para receber um valor X entre os valores únicos encontrados em cada coluna
list_to_fix = ['BytePairsReoccuringUpFlow.FeatureBytePairsValue', 'BytePairsReoccuringDownFlow.FeatureBytePairsValue', 'First4BytesHashUpFLow.FeatureHashValue', 'First4BytesHashDownFlow.FeatureHashValue']
for columns in list_to_fix:
    temporary = df[columns].unique()
    df[columns] = df[columns].replace( temporary , [ i for i in range(0,len(temporary))] ).astype('float64')
```

```python
# df['ApplicationProtocolName'] = df['ApplicationProtocolName'].map({'53_UDP_Dns': 0, '80_TCP_HTTP': 1}).astype(int)
```

```python
# Criação de colunas adicionais ao vetor de características
df['sum'] = df.sum(axis=1, numeric_only=True)
df['mean'] = df.mean(axis=1, numeric_only=True)
df['var'] = df.var(axis=1, numeric_only=True)
df['std'] = df.std(axis=1, numeric_only=True)
```
